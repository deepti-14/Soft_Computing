```python
import numpy as np
import random
from math import *
```

```python
# Initialization
N_pop = 500
N_elite = 5
max_iter = 10
pc = 0.05
pm = 0.1
```

```python
# for objective function 1
def calculateFitness1(curr):
    f = 0
    total_w = 0
    for j in range (N_pop):
        total_w += random.random()
    w = random.random()/total_w
    f = (pow(curr,2))*w
    return f

# for objective function 2
def calculateFitness2(curr):
    f = 0
    total_w = 0
    for j in range (len(population)):
        total_w += random.random()
    w = random.random()/total_w
    f = (pow(curr-2,2))*w
    return f
```

```python
def calculateProbability(fitness):
    den = 0
    f_min = min(fitness)
    for i in range(len(population)):
        den += (fitness[i]-f_min)

        prob = []
        for i in range(len(population)):
            if den == 0:
                den = 1
            p = (fitness[i]-f_min)/den
    prob.append(p)
```

```python
def selection1(population):
    fitness1 = []
    for i in range(len(population)):
        f = calculateFitness1(population[i])
        fitness1.append(f)
    prob = calculateProbability(fitness1)
    selected_pairs = []

    for i in range(len(population)//2):
        # select pair according to probability
        pair = random.choices(population, weights=prob, k=2)
        selected_pairs.append(pair)

    return selected_pairs
```

```python
def selection2(population):
    fitness2 = []
    for i in range(len(population)):
        f = calculateFitness2(population[i])
        fitness2.append(f)
    prob = calculateProbability(fitness2)
    selected_pairs = []

    for i in range(len(population)//2):
        pair = random.choices(population, weights=prob, k=2)
        selected_pairs.append(pair)

    return selected_pairs
```

```python
def crossover(pair):
    neg = False
    if pair[0] < 0 or pair[1] < 0:
        neg = True
    parent1 = abs(pair[0])
    parent2 = abs(pair[1])
    # converting to list for crossover
    parent1 = list(map(int, str(parent1)))
    parent2 = list(map(int, str(parent2)))

    end = min(len(parent1), len(parent2))
    crossover_point = random.randint(0, end)

    for i in range(crossover_point):
        parent1[i], parent2[i] = parent2[i], parent1[i]
    # converting back into integer
    child1 = int("".join(map(str, parent1)))
    child2 = int("".join(map(str, parent2)))

    # checking if any value was negative
    if random.uniform(0, 1) > 0.5 and neg == True:
        child1 = -(child1)
        child2 = -(child2)
    return [child1, child2]
```

```python
def mutation(val):
    neg = False
    if val < 0:
        neg = True
    val = abs(val)

    val = list(map(int, str(val)))
    end = len(val)
    mutation_point = random.randint(0, end-1)

    val[mutation_point] = random.randint(0,10)
    val = int("".join(map(str, val)))

    if random.uniform(0, 1) > 0.5 and neg == True:
        val = -(val)
    return val
```

```python
# remove n_elite strings
def remove(population):
    added= population
    if len(population) >= N_elite:
        # deleting N_elite strings
        to_delete = set(random.sample(range(len(population)), N_elite))
        deleted = [x for i,x in enumerate(population) if not i in to_delete]

    # adding N_elite strings from inital pop
    added = deleted + random.choices(initial_pop, k=N_elite)

    return added
```

### FOR OBJECTIVE FUNCTION X^2

```python
population = []
for i in range (N_pop):
        k = random.randint(-1000,1000)
        population.append(k)

initial_pop = population
for i in range(max_iter):
    print("Iteration ", i+1)
    selected_pairs1= selection1(population)
    offsprings= []
    for i in range(len(selected_pairs1)):
        # applying crossover
        if random.uniform(0,1) > pc:
            offspring = crossover(selected_pairs1[i])
            offsprings.append(offspring[0])
            offsprings.append(offspring[1])

#     print("Offsprings after crossover:", offsprings)
    new_pop = []
    for i in range(len(offsprings)):
        if random.uniform(0,1) > pm:
            new_pop.append(mutation(offsprings[i]))

#     print("\nOffsprings after mutation:", new_pop)

    new_pop = remove(new_pop)
    population = new_pop

    print("\nNew population: ", new_pop)

ans = max(max(population), abs(min(population)))

print("\nBest value of x:", sqrt(ans))
print("f1(x)= ", ans)
```
